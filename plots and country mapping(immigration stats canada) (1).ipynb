```python
# top 15 countries based on Total immigrant population
df_top15 = df_con.sort_values(['Total'], ascending=False, axis=0).head(15)
df_top15
```

```python
# new dataframe which contains the aggregate for each decade

# list of all years in decades 80's and 90's
years_80s = list(map(str, range(1980, 1990))) 
years_90s = list(map(str, range(1990, 2000))) 
years_00s = list(map(str, range(2000, 2010))) 

# slice the original dataframe df_con to create a series for each decade
df_80s = df_top15.loc[:, years_80s].sum(axis=1) 
df_90s = df_top15.loc[:, years_90s].sum(axis=1) 
df_00s = df_top15.loc[:, years_00s].sum(axis=1)

# merge the two series into a new data frame
new_df = pd.DataFrame({'1980s': df_80s, '1990s': df_90s, '2000s':df_00s}) 

print(new_df)
new_df.describe()
```

```python
new_df.plot(kind='box', figsize=(10, 6))

plt.title('Immigration from top 15 countries for decades 80s, 90s and 2000s')

plt.show()
```

```python
df_tot = pd.DataFrame(df_con[years].sum(axis=0))
# change the years to type float (useful for regression later on)
df_tot.index = map(float,df_tot.index)
df_tot.reset_index(inplace = True)
df_tot.columns = ['year', 'total']
df_tot.head()
```

```python
df_tot.plot(kind='scatter', x='year', y='total', figsize=(10, 6), color='darkblue')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.show()
```

```python
x = df_tot.year      # year on x-axis
y = df_tot.total     # total on y-axis
fit = np.polyfit(x, y, deg=1)

fit
```

```python
df_tot.plot(kind='scatter', x='year', y='total', figsize=(10, 6), color='darkblue')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

# plot line of best fit
plt.plot(x,fit[0]*x + fit[1], color='red') 
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(2000, 150000))

plt.show()

# Print out the line of best fit
'No. Immigrants = {0:.0f} * Year + {1:.0f}'.format(fit[0], fit[1]) 
```

```python
df_con_t=df_con[years].transpose() # transposed dataframe
df_con_t.index = map(float,df_con_t.index)
df_con_t.index.name = 'Year'
# reset index to bring the Year in as a column
df_con_t.reset_index(inplace = True)
df_con_t.head(3)
```

```python
norm_brazil = (df_con_t.Brazil - df_con_t.Brazil.min()) / (df_con_t.Brazil.max() - df_con_t.Brazil.min())
norm_argentina = (df_con_t.Argentina - df_con_t.Argentina.min()) / (df_con_t.Argentina.max() - df_con_t.Argentina.min())

norm_brazil
```

```python
# Brazil
ax0 = df_con_t.plot(kind='scatter',
                    x='Year',
                    y='Brazil',
                    figsize=(14, 8),
                    alpha=0.5,                  # transparency
                    color='green',
                    s=norm_brazil * 2000 + 10,  # pass in weights 
                    xlim=(1975, 2015)
                   )

# Argentina
ax1 = df_con_t.plot(kind='scatter',
                    x='Year',
                    y='Argentina',
                    alpha=0.5,
                    color="blue",
                    s=norm_argentina * 2000 + 10,
                    ax = ax0
                   )

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from Brazil and Argentina from 1980 - 2013')
ax0.legend(['Brazil', 'Argentina'], loc='upper left', fontsize='x-large')
```
